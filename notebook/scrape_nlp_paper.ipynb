```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
def get_nlp_papers_from2014_to2019(URL):
    resp = requests.get(URL)
    soup = BeautifulSoup(resp.content, 'html.parser')

    start_flag = False
    title_flag = False

    start_line = '    ä¸€èˆ¬ç™ºè¡¨'
    end_line = '<h2 id="author_list">'

    no_pattern = r'<span id="(.*[0-9]+-[0-9]+)">'
    title_pattern = r'<span[=""]* class="title".*>'
    auther_pattern = r' *([ ^]*â—‹.*)'

    no = False
    title = False
    auther = False
    l = []

    for line in soup.prettify().split("\n"):
        if start_flag: #è«–æ–‡ä¸€è¦§ãŒå§‹ã¾ã£ãŸã‚‰ã“ã“ã«å…¥ã‚‹

            no_m = re.search(no_pattern, line)
            title_m = re.search(title_pattern, line)
            auther_m = re.search(auther_pattern, line)

            if (not no) and (not title) and no_m:
                no = no_m.group(1)

            elif no and title_m:
                title_flag = True

            elif title_flag:
                title = line.strip()
                title_flag = False

            elif no and title and auther_m:
                auther = auther_m.group(1)

                l.append([no, title, auther])
                no = False
                title = False
                auther = False

            else:
                pass

            if end_line in line: #ç´¢å¼•ãŒå§‹ã¾ã£ãŸã‚‰çµ‚ã‚ã‚Š
                print("end")
                break
        elif start_line in line: 

            print("start")
            start_flag = True

    
    return pd.DataFrame(l, columns=["id", "title", "auther"])
```

```python
# NLP2014-2019å¹´ã‚’æŠœãå‡ºã™

df_nlp_2014_2019 = ""
for i in range(2014, 2020):
    print(i, "å¹´  ã‚¹ã‚¯ãƒ¬ã‚¤ãƒ”ãƒ³ã‚°é–‹å§‹")
    URL = "http://www.anlp.jp/proceedings/annual_meeting/%i/index.html" % (i)
    df_ = get_nlp_papers_from2014_to2019(URL)
    df_["year"] = i
    
    if len(df_nlp_2014_2019) != 0:
        df_nlp_2014_2019 = pd.concat([df_nlp_2014_2019, df_])
    else:
        df_nlp_2014_2019 = df_.copy()
        
df_nlp_2014_2019.head()
```

```python
def make_github_issue(title, body=None, labels=None):
    '''Create an issue on github.com using the given parameters.'''
    # Our url to create issues via POST
    url = 'https://api.github.com/repos/%s/%s/issues' % (REPO_OWNER, REPO_NAME)
    # Create an authenticated session to create the issue
    session = requests.Session()
    session.auth = (USERNAME, PASSWORD)
    # Create our issue
    issue = {'title': title,
             'body': body,
             'labels': labels}
    # Add the issue to our repository
    r = session.post(url, json.dumps(issue))
    if r.status_code == 201:
        print ('Successfully created Issue {0:s}'.format(title))
    else:
        print ('Could not create Issue {0:s}'.format(title))
        print ('Response:', r.content)
```
